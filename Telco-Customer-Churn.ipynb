```python
plt.figure(figsize = (3,4)) 
ax = sns.countplot(x = dataset['SeniorCitizen'], data = dataset)
ax.bar_label(ax.containers[0], fontsize = 9)
plt.title("Churn by Senior Citizen")
plt.show()
```

```python
# Preprocessing: Calculate percentage distribution for each group
grouped = dataset.groupby(['SeniorCitizen', 'Churn']).size().unstack()  # Create a pivot table
grouped_percentage = grouped.div(grouped.sum(axis=1), axis=0) * 100  # Convert to percentages

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(5, 4))
grouped_percentage.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])

# Add percentage labels
for i, bars in enumerate(ax.containers):
    for bar in bars:
        height = bar.get_height()
        if height > 0:  # Only display non-zero percentages
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f'{height:.1f}%',
                ha='center', va='center', color='black', fontsize=8
            )

# Add titles and labels
ax.set_title('Churn by Senior Citizen', fontsize=14)
ax.set_xlabel('Senior Citizen', fontsize=12)
ax.set_ylabel('Percentage', fontsize=12)
ax.legend(title='Churn', fontsize=10, bbox_to_anchor = (0.9 , 0.9))
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

```python
# List of columns to visualize
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set the number of rows and columns for the subplots
num_cols = 3  # Number of columns in the grid
num_rows = (len(columns) + num_cols - 1) // num_cols  # Calculate rows to fit all subplots

# Create a figure with subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 4), constrained_layout=True)
axes = axes.flatten()  # Flatten the 2D axes array for easier iteration

# Plot each column as a count plot
for i, column in enumerate(columns):
    sns.countplot(data=dataset, x=column, ax=axes[i], palette="Set2", hue = dataset['Churn'])
    axes[i].set_title(f'Distribution of {column}', fontsize=12)
    axes[i].set_xlabel(column, fontsize=10)
    axes[i].set_ylabel('Count', fontsize=10)
    axes[i].tick_params(axis='x')

# Hide any unused subplots
for j in range(len(columns), len(axes)):
    axes[j].axis('off')


plt.show()
```

From the provided graphs, here's an analysis of the insights for each of the columns represented:

---

#1. PhoneService
- The majority of customers have phone service.
- Customers with phone service are more likely to churn compared to those without it.

---

#2. MultipleLines**
- Most customers either have no additional phone lines or only a single line.
- Churn is higher among customers with additional phone lines compared to those without.

---

#3. InternetService
- A significant number of customers use "Fiber optic" internet, followed by "DSL."
- Customers with "Fiber optic" internet exhibit higher churn rates compared to "DSL" users.
- Those without internet service show the least churn.

---

#4. OnlineSecurity
- Many customers do not have online security services.
- Churn is significantly higher among customers who lack online security compared to those with the service.

---

#5. OnlineBackup
- Similar to OnlineSecurity, a large number of customers do not have online backup services.
- Customers without online backup services are more likely to churn compared to those with it.

---

#6. DeviceProtection
- Many customers do not have device protection services.
- Churn rates are higher for customers who do not have device protection compared to those who do.

---

#7. TechSupport
- A significant portion of customers lack technical support services.
- Customers without tech support are more likely to churn, indicating the importance of offering support.

---

#8. StreamingTV
- A large number of customers either subscribe to streaming TV services or do not use them due to a lack of internet service.
- Customers with streaming TV services have a slightly higher churn rate than those without.

---

#9. StreamingMovies
- The pattern is similar to StreamingTV, where many customers either subscribe or do not use streaming movies due to no internet service.
- Churn is higher among customers with streaming movies subscriptions.
